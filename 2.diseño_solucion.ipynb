**PROBLEMA INICIAL**

Se requiere identificar el fraude que se comete en una entidad financiera mediante el uso de tarjetas de crédito.


**¿Que se quiere lograr?**
- Predecir potenciales usuarios que podrian cometer el fraude en un futuro, de esta forma la entidad definira a que usuarios les aprueba tarjetas, creditos, e.t.c. y cuales usuarios les rechaza su solicitud.

- Aumentar la seguridad de la sociedad y velar por la buena imagen de la entidad financiera



**Retos del problema**

- La variable *class*, que determina en la base de datos historica que usuarios han cometido fraudes y cuales no, se encuentra muy desbalanceada, por lo cual esto puede generar problemas en el modelo de predicción, se deben utilizar tecnicas de balanceo de clases en este campo para velar por la presicion de las predicciones.

**Zoom analítico**

- Se entrenara el modelo semestralmente para evaluar el rendimiento y la presicion de sus predicciones
- Los datos recopilados mensualmente se almacenaran mensualmente en la base de datos del año respectivo para que de esta forma esta sirva como insumo de entrenamiento del modelo
